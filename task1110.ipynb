```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

belize_listings = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings)
```

Q.1 a. A relationship is linear when a change in one variable is associated with a proportional change in the other variable. In a monotonic relationship, the variables tend to change together, but not necessarily at a constant rate. A ranked relationship is based on the ranked values for each variable rather than the raw data.

```python
x= 'calculated_host_listings_count' 
y= 'availabilty_365'
fig, ax = plt.subplots()

ax.plot(belize_df['calculated_host_listings_count'], belize_df['availability_365'], marker = "o", linestyle='None', color= 'red')

# Set the x-axis label
ax.set_xlabel('Calculated Host Listings Count')

# Set the y-axis label 
ax.set_ylabel('Availability_365')

# Set the title
ax.set_title("Relationship between Calculated Host Listings Count & Availability 365")

plt.grid()

plt.show()

#calculate the correlation coefficient

np.corrcoef(belize_df['calculated_host_listings_count'],belize_df['availability_365'])

#There is a negligible relationship between the two, with plots scattered all over

```
