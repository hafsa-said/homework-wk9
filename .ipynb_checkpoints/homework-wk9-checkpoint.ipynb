# Monty Hall Problem

The Monty Hall problem is based on a game show, where it decides whether you have a greater probability of getting a correct answer if you switch your answers after a wrong answer. This is because when you start, you have an equal 1/3rd chance of winning at each door, but once another door has been exposed, when you switch your doors, you move over to the other 2/3 probability of you winning rather than the 1/3rd you originally started with. Especially, because Monty will never expose the door with the car behind it, by switching you increase the likelihood that you have chosen the correct answer.
